```
#Cree la clase finanzas según el diagrama
class Finance:
  def __init__(self, checkout):
    self.checkout = checkout
  def investment(self, money):
    self.checkout = self.checkout - money
  def injection(self, money):
    self.checkout = self.checkout + money
```

```
#Copie y pegue la creación de la clase máquina y modifique el código
class Machine:
  def __init__(self, name, type_machine, serie, location, brand, supplier, status, use_hours, number_maintenances):
    if (type_machine in type_machine_list) and (location in location_list) and (brand in brand_list) and (supplier in supplier_list) and (status in status_list):
      self.name = name
      self.type_machine = type_machine
      self.serie = serie
      self.location = location
      self.brand = brand
      self.supplier = supplier
      self.status = status
      self.use_hours = use_hours
      self.number_maintenances = number_maintenances
    else:
      print("This object can't be create. Be sure every item is part of the options.")
      # self.__del__()    

  def use(self, hours):
    self.use_hours += hours
    maintenance_hours = {'Caminadora': 50, 'Eliptica': 35, 'Escaladora': 70, 'Pesas': 120, 'Multifuncional': 40}
    if (self.type_machine in maintenance_hours) and (self.use_hours > maintenance_hours[self.type_machine]) and self.status == "Disponible":
      self.status = "Nec_repa"
 
  def fix(self, finance):
    maintenance_cost = {'Eliptica KMG': 140, 'Caminadora KMG': 12, 'Caminadora MX': 132, 'Kola Hatha': 26}   
    type_machine_brand = self.type_machine+" "+self.brand
    if (type_machine_brand in maintenance_cost):
      if maintenance_cost[type_machine_brand] < finance.checkout:
        finance.investment(maintenance_cost[type_machine_brand])
        self.number_maintenances += 1
        self.status = "Disponible"
        print("{}: Maintenance realized. Fix cost: {}. New checkout: {}. Number of maintenances: {}" .format(self.name,maintenance_cost[type_machine_brand],finance.checkout,self.number_maintenances))
      else:
        print("{}: Maintenance not realized. There is not enough money to do this fix. Fix cost: ${}. Checkout: ${}." .format(self.name,maintenance_cost[type_machine_brand], finance.checkout))
    else:
      print("{}: Maintenance not realized. There is not info about this machine fix cost." .format(self.name))
    
```
